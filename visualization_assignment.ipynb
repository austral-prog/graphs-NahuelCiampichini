```python
# 1) grafico de lineas
import matplotlib.pyplot as plt
import numpy as np

dias = np.arange(1, 8)
temp = [22, 24, 19, 23, 25, 27, 26]

plt.figure()
plt.plot(dias, temp, color='blue', marker='o', linestyle='--', label='Temperatura')
plt.title('Temperatura diaria durante una semana')
plt.xlabel('Dia')
plt.ylabel('Temperatura (°C)')
plt.legend()
plt.grid(True)

plt.show()
```

Esto es un grafico de lineas que muestra las variaciones de la temperatura durante la semana.

```python
# 2) Grafico de barras
import matplotlib.pyplot as plt
import numpy as np

cate = ['A', 'B', 'C', 'D']
val = [15, 22, 8, 19]

plt.figure()
plt.bar(cate, val, color='green')
plt.title('Ventas por categoría')
plt.xlabel('Categoría')
plt.ylabel('Ventas p/u')
plt.show()
```

Esto es un grafico de barras que compara diferentes categorias usando las ventas por unidad de cada una.

```python
# 3) Diagrama de dispersion
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = 2 * x + np.random.normal(0, 0.1, 50)

plt.figure()
plt.scatter(x, y, color='red', marker='x', label='Puntos de datos')
plt.title('Diagrama de dispersion de x, y')
plt.xlabel('Valor de x')
plt.ylabel('Valor de y')
plt.legend()
plt.show()
```

Este es un diagrama de dispersion que muestra la relacion de la variable x con la variable y

```python
#4 histograma
import matplotlib.pyplot as plt
import numpy as np

datos = np.random.randn(1000)

plt.figure()
plt.hist(datos, bins=30, color='purple', edgecolor='black')
plt.title('Distribucion de datos aleatorios')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

esto es un histograma que muestra la frecuencia de valores.

```python
# 5) Grafico de torta
import matplotlib.pyplot as plt
import numpy as np

labels = ['Categoria A', 'Categoria B', 'Categoria C', 'Categoria D']
sizes = [25, 35, 20, 20]

plt.figure()
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Proporcion de las categorias')
plt.axis('equal')

plt.show()
```

Este es un grafico de torta que muestra las proporciones de cada categoria sobre el total
